```python
import pandas as pd

summary = pd.read_csv("../data/processed/kpi_worldcup.csv")
summary.tail()
```

```python
last5 = summary.tail(5)

gpm = last5["goals_per_match"].mean()

matches_2026 = 104
teams_2026 = 48

proj_goals = gpm * matches_2026
proj_goals_team = proj_goals / teams_2026

projection = {
    "Goals_per_match": gpm,
    "Total_goals": proj_goals,
    "Goals_per_team": proj_goals_team
}

projection
```

```python
pd.DataFrame([projection]).to_csv("../data/processed/projection_2026.csv", index=False)
```
