```python
import sys
from pathlib import Path
sys.path.append(str(Path("..").resolve()))

import pandas as pd

matches = pd.read_csv("../data/processed/matches_clean.csv")
```

```python
HOST_BY_YEAR = {
    2018: "Russia",
    2022: "Qatar",
}
```

```python
matches["host"] = matches["year"].map(HOST_BY_YEAR)
matches = matches.dropna(subset=["host"])
```

```python
host_games = matches[
    (matches["home_team_name"] == matches["host"]) |
    (matches["away_team_name"] == matches["host"])
].copy()
```

```python
def host_result(row):
    if row["home_team_name"] == row["host"]:
        return row["home_team_goals"] > row["away_team_goals"]
    else:
        return row["away_team_goals"] > row["home_team_goals"]

host_games["host_win"] = host_games.apply(host_result, axis=1)
```

```python
host_games.groupby("year")["host_win"].mean()
```

```python
attendance = matches.groupby("year")["attendance"].mean().reset_index()
attendance
```

```python
attendance.plot(x="year", y="attendance", title="Average Attendance Evolution")
```

```python
goals = matches.groupby("year")["total_goals"].mean().reset_index()

merged = attendance.merge(goals, on="year")

merged.corr()
```
